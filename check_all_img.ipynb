```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.models import load_model


model = load_model('/mnt/d/Coding/tensorflowProjects/tf217/final_model.keras')

test_dir = '/mnt/d/Coding/tensorflowProjects/tf217/test'
class_labels = {0: 'DR', 1: 'No_DR'}
label_map = {'DR': 0, 'NO_DR': 1}  


def load_and_preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array


total_images = 0
correct_predictions = 0


for label_name, label_value in label_map.items():
    folder_path = os.path.join(test_dir, label_name)
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        img_array = load_and_preprocess_image(img_path)
        
        
        prediction = model.predict(img_array)
        predicted_class = int(prediction[0, 0] > 0.5)  # 0 for 'DR', 1 for 'No_DR'
        
        
        if predicted_class == label_value:
            correct_predictions += 1
        total_images += 1


accuracy = (correct_predictions / total_images) * 100
print(f"Total images: {total_images}")
print(f"Correct predictions: {correct_predictions}")
print(f"Accuracy on test set: {accuracy:.2f}%")
```
